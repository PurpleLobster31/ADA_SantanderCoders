## Funções

A sintaxe se dá de tal maneira:
```
def <identificador da func> (<parametros da func>):
    <codigo identado>
    return <o que pode ser retornado>
```

````python
def is_float(x):
    if type(x) == float:
        print("É float!")
    else:
        print("Não é float!")
````

````python
print(is_float(4))
````

````python
print(is_float(5.03))
````

````python
def is_float_bool(x):
    if type(x) == float:
        return True
    else:
        return False
````

````python
val = 4.75
if is_float_bool(val):
    print("É float!")
````

## Coleções Aninhadas

Listas, tuplas, dicionários e outros podem ser aninhados!

````python
# Lista de tuplas

fruta1 = (640, 'Morango', 25.0)
fruta2 = (200, 'Banana', 37.50)
fruta3 = (487, 'Abacaxi', 70.0)
````

````python
lista_de_frutas = [fruta1, fruta2, fruta3]
print(lista_de_frutas)
````

````python
lista_de_frutas.sort()
lista_de_frutas
````

````python
# Aninhando dicionários e listas

notas = {
    'Python': {2020: [9.5, 8.0, 10.0, 5.0, 6.0, 6.0],
               2021: [10.0, 7.5, 3.5, 8.5, 10.0, 10.0]
    },
    'Redes Neurais':{2020: [10.0, 7.5],
                     2021: [2.5, 3.5, 8.0],
                     2022: [7.75, 8.2, 3.7 ]
        }
}
````

## List Comprehension

A imagem abaixo representa bem:
    

Sintaxe:
```
lista = [expressao for variavel_local in objeto]
```

Equivalente a:
```
lista = []
for variavel_local in objeto:
    lista.append(expressao variavel_local)
```

````python
quad = [x**2 for x in range(0,11)]
quad
````

### Condições em compreensão de lista

````python
metade_pares = [x/2 for x in range(0,100) if x%2==0]
````

````python
print(metade_pares)
````

````python
metade_par_dobro_impar = [x/2 if x%2==0 else x*2 for x in range(0,100)]
print(metade_par_dobro_impar)
````

````python

ls3 = ["Aqui era zero" if x == 0 else "par" if x%2==0 else "ímpar" for x in range(0,100)]
print(ls3)
````

### Comprehensions com iteracão paralela em diferentes coleções

```
    lista = [expressao for <var1, var2, ...> in zip(<colec1, colec2, ...>)]
```

````python
l1 = list(range(0,10))
l2 = [x/10 for x in range(0,10)]
````

````python
resultado = list(zip(l1,l2))
resultado
````

### Comprehensions aninhados

```
lista = [expressao for var_local1 in obj1 if cond1
                    for var_local2 in obj2 if cond2]
```

Equivalente a:
```
lista = []
for var_local1 in obj1:
    if cond1:
        for var_local2 in obj2:
            if cond2:
                lista.append(expressao)
```

````python
A = ['a', 'b']
B = [10,20,30]
````

````python
prod_cart = {(a,b) for a in A for b in B}
print(prod_cart)
````

````python
prod_cart_dict = {a:b for a in A for b in B}
print(prod_cart_dict)
````
